```python
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
style.use('ggplot')
x=[1,5,4]
y=[23,45,32]
x2=[3,7,9]
y2=[8,32,6]
plt.plot(x,y,label="First",linewidth=3,marker='*')
plt.plot(x2,y2,label="second",linewidth=4,marker='s')
plt.legend(loc='upper left')
plt.show()
```

```python
student_performance=["Excellent","Good","Average","Poor"]
student_values=[15,25,12,8]
plt.pie(student_values,labels=student_performance)
plt.show()
```

```python

student_performance=["Excellent","Good","Average","Poor"]
student_values=[15,25,12,8]
plt.pie(student_values,labels=student_performance,startangle=90)
plt.show()
```

```python
student_performance=["Excellent","Good","Average","Poor"]
student_values=[15,25,12,8]
plt.pie(student_values,labels=student_performance,explode=[0.2,0,0.08,0])
plt.show()
```
